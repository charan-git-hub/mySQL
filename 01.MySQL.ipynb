## Types of Database Languages

1. **Data Definition Language (DDL)**
   - DDL is used to define the structure of a database. It includes commands for creating, altering, dropping, and modifying database objects.
   - **Common DDL Commands**:
     - `CREATE`: Used to create new tables or databases.
     - `ALTER`: Used to modify existing database structures.
     - `DROP`: Used to delete tables or databases.
     - `TRUNCATE`: Used to remove all records from a table but keep the table structure intact.

2. **Data Manipulation Language (DML)**
   - DML is used for manipulating data within the database. It includes commands for inserting, updating, deleting, and retrieving data.
   - **Common DML Commands**:
     - `SELECT`: Used to query data from tables.
     - `INSERT`: Used to add new records to tables.
     - `UPDATE`: Used to modify existing records.
     - `DELETE`: Used to remove records from tables.

3. **Data Control Language (DCL)**
   - DCL is used to control access to data within the database. It includes commands for granting and revoking permissions.
   - **Common DCL Commands**:
     - `GRANT`: Gives specific privileges to users.
     - `REVOKE`: Removes specific privileges from users.

4. **Transaction Control Language (TCL)**
   - TCL is used to manage transactions in the database, ensuring data integrity.
   - **Common TCL Commands**:
     - `COMMIT`: Saves all changes made in the current transaction.
     - `ROLLBACK`: Undoes changes made in the current transaction.
     - `SAVEPOINT`: Sets a point in a transaction to which you can later roll back.

Each of these languages serves a specific purpose in the management and manipulation of databases, making it easier for users to interact with and maintain their data.

## RDBMS Terminology

### 1. Table
- **Definition**: A table is a collection of related data organized in rows and columns. Each table represents a specific entity (e.g., customers, orders).
- **Example**: A `customers` table might have columns for `customer_id`, `name`, `email`, and `phone_number`.#

## 2. Row (Record)
- **Definition**: A row, also known as a record or tuple, represents a single entry in a table. Each row contains data for the columns defined in the table.
- **Example**: In a `customers` table, a row might represent one customer, such as `(1, 'John Doe', 'john@example.com', '123-456-7890#')`.

## 3. Column (Field)
- **Definition**: A column, also known as a field or attribute, represents a specific characteristic of the entity defined by the table. Each column has a defined data type (e.g., integer, varchar).
- **Example**: In a `customers` table, columns could include `customer_id`, `name`, `email`, and `phone_#number`.

## 4. Primary Key
- **Definition**: A primary key is a unique identifier for each row in a table. It ensures that no two rows can have the same value in this column.
- **Example**: The `customer_id` column in the `customers` table can be a #primary key.

## 5. Foreign Key
- **Definition**: A foreign key is a column (or set of columns) in one table that refers to the primary key of another table, establishing a relationship between the two tables.
- **Example**: In an `orders` table, a `customer_id` column could serve as a foreign key referencing the `customer_id` in the `c#ustomers` table.

## 6. Index
- **Definition**: An index is a database object that improves the speed of data retrieval operations on a table. It creates a data structure that allows for faster searches.
- **Example**: An index on the `email` column of the `customers` table can speed up queries that search for# customers by email.

## 7. Schema
- **Definition**: A schema is the overall structure of the database, including the tables, columns, relationships, constraints, and other database objects.
- **Example**: A database schema might define a `sales` schema containing tables for `customers`, `#orders`, and `products`.

## 8. Query
- **Definition**: A query is a request for data or information from a database. Queries are typically written in SQL (Structured Query Language).
- **Example**: A SQL query like `SELECT * FROM customers WHERE name = 'John Doe';` retrieves all information about #the customer named John Doe.

## 9. View
- **Definition**: A view is a virtual table that presents data from one or more tables. It does not store the data itself but provides a way to simplify complex queries.
- **Example**: A view named `customer_orders` might combine data from the `customers` and `orders` tables to show customer name#s alongside their order details.

## 10. Transaction
- **Definition**: A transaction is a sequence of one or more SQL operations that are executed as a single unit. Transactions ensure data integrity and consistency.
- **Example**: A transaction may involve inserting an order and updating the customer's order history; both opera#tions must succeed or fail together.

## 11. Normalization
- **Definition**: Normalization is the process of organizing data in a database to reduce redundancy and improve data integrity. It involves dividing large tables into smaller ones and defining relationships between them.
- **Example**: Instead of storing customer information with every order, normalization might involve separating customers and orders into distinct ta#bles and linking them with foreign keys.

## 12. Denormalization
- **Definition**: Denormalization is the process of intentionally introducing redundancy into a database to improve read performance. It can simplify queries at the expense of increased storage and potential data anomalies.
- **Example**: Storing customer information in the `orders` table alongside order details to reduce the need for complex joins when querying orders.

## 13. Constraint
- **Definition**: A constraint is a rule applied to a column in a table that restricts the types of data that can be inserted. Constraints help maintain data integrity.
- **Example**: A `NOT NULL` constraint on the `name` column in the `customers` table ensures that every customer must have a name.

## 14. Stored Procedure
- **Definition**: A stored procedure is a precompiled collection of one or more SQL statements that can be executed as a single unit. It can include control flow statements and variables.
- **Example**: A stored procedure might be created to process orders, automatically updating inventory and generating invoices.

## 15. Trigger
- **Definition**: A trigger is a set of instructions that automatically executes in response to specific events on a particular table, such as insertions, updates, or deletions.
- **Example**: A trigger can be set to automatically update an inventory table whenever a new order is placed.

Understanding these terms is crucial for effectively working with relational databases and leveraging their capabilities to manage data efficiently.

## Comments in MySQL

- In MySQL, you can add comments to your SQL code to explain or annotate specific lines or sections. This is helpful for documentation and makes your code more understandable to others (or yourself in the future).

- In MySQL, comments are used to provide explanations or annotations within SQL code. Comments have no impact on the execution of the SQL statements; they are ignored by the database engine. Comments are useful for documenting the code, making it more readable, and explaining the purpose of certain sections or lines. MySQL supports two types of comments:

#### 1. Single-line Comments

You can use two dashes (`--`) followed by a space or the hash symbol (`#`) to create a single-line comment.

Single-line comments start with the `--` (double hyphen) syntax and continue to the end of the line. Anything after `--` is treated as a comment.

**Example:**
```sql
-- This is a single-line comment
SELECT * FROM employees;  -- Fetch all records from the employees table

# This is another single-line comment
SELECT * FROM customers;  # Fetch all records from the customers table


#### 2. Multi-Line Comments:

For multi-line comments, you can enclose your comments within `/*` and `*/`.

Multi-line comments start with `/*` and end with `*/`. Everything between `/*` and `*/` is treated as a comment, and it can span multiple lines.

**Example:**
```sql
/*
This is a multi-line comment.
It can span multiple lines.
*/
SELECT * FROM orders;

/*
Retrieve all orders placed in the last month
This query filters by the order date
*/
SELECT * FROM orders WHERE order_date >= NOW() - INTERVAL 1 MONTH;

# MySQL Data Types

MySQL supports various data types that can be used to define the type of data that can be stored in a table. Understanding these data types is crucial for designing efficient databases. Below is a detailed explanation of the primary categories of data types in MySQL.

## 1. Numeric Types

Numeric data types are used to store numbers. 

### INT
- **Description**: The `INT` data type stores standard integer values. 
- **Range**: 
  - Signed: -2,147,483,648 to 2,147,483,647
  - Unsigned: 0 to 4,294,967,295
- **Usage**: Commonly used for whole numbers, such as IDs or counts.

### DECIMAL(M, D)
- **Description**: The `DECIMAL` data type stores fixed-point numbers. 
- **Parameters**: 
  - `M` specifies the maximum number of digits (precision).
  - `D` specifies the number of digits to the right of the decimal point (scale).
- **Usage**: Useful for precise calculations, such as storing currency values. For example, `DECIMAL(10, 2)` can store numbers up to 99999999.99.

### FLOAT
- **Description**: The `FLOAT` data type stores floating-point numbers. 
- **Storage**: Requires 4 bytes of storage.
- **Usage**: Used for approximate numeric values, where precision is not critical, such as scientific calculations.

## 2. String Types

String data types are used to store text and character data.

### CHAR(M)
- **Description**: The `CHAR` data type stores fixed-length strings.
- **Length**: Can store up to 255 characters.
- **Behavior**: If the string is shorter than `M`, it is padded with spaces.
- **Usage**: Suitable for strings of consistent length, such as postal codes.

### VARCHAR(M)
- **Description**: The `VARCHAR` data type stores variable-length strings. 
- **Length**: Can store up to 65,535 characters.
- **Storage**: Only uses as much storage as needed, plus 1 or 2 bytes to store the length.
- **Usage**: Ideal for strings of varying lengths, such as names or email addresses.

### TEXT
- **Description**: The `TEXT` data type stores variable-length strings.
- **Length**: Can store up to 65,535 characters.
- **Usage**: Useful for larger text data, such as descriptions or articles, where the length is unpredictable.

## 3. Date and Time Types

Date and time data types are used to store date and time values.

### DATE
- **Description**: The `DATE` data type stores dates.
- **Format**: `YYYY-MM-DD`
- **Range**: From `'1000-01-01'` to `'9999-12-31'`.
- **Usage**: Ideal for storing birthdates, event dates, etc.

### DATETIME
- **Description**: The `DATETIME` data type stores both date and time.
- **Format**: `YYYY-MM-DD HH:MM:SS`
- **Range**: From `'1000-01-01 00:00:00'` to `'9999-12-31 23:59:59'`.
- **Usage**: Useful for tracking timestamps, such as when a record is created or modified.

### TIMESTAMP
- **Description**: The `TIMESTAMP` data type stores date and time, similar to `DATETIME`, but with additional features.
- **Range**: From `'1970-01-01 00:00:01'` UTC to `'2038-01-19 03:14:07'` UTC.
- **Usage**: Automatically updates to the current timestamp when a record is created or modified. Ideal for tracking changes in records.

## 4. Boolean Type

### BOOL (or BOOLEAN)
- **Description**: A synonym for TINYINT(1).
- **Values**: Typically used to represent true (1) or false (0) values.
- **Usage**: Ideal for flags or binary state indicators, such as whether a user is active or not.

## 5. Binary Types

Binary data types are used to store binary data.

### BINARY(M)
- **Description**: The `BINARY` data type stores fixed-length binary strings.
- **Length**: Can store up to 255 bytes.
- **Behavior**: If the binary string is shorter than `M`, it is padded with zero bytes.
- **Usage**: Suitable for storing fixed-length binary data, such as hashes.

### VARBINARY(M)
- **Description**: The `VARBINARY` data type stores variable-length binary strings.
- **Length**: Can store up to 65,535 bytes.
- **Storage**: Only uses as much storage as needed, plus 1 or 2 bytes to store the length.
- **Usage**: Ideal for storing binary data of varying lengths, such as images or files.

## Choosing the Right Data Type

Choosing the right data type is essential for optimizing storage, performance, and data integrity in your database. Consider the nature of the data, the expected range of values, and any specific requirements when selecting data types for your tables.


# MySQL Data Types

MySQL supports various data types that can be used to define the type of data that can be stored in a table. Below is a detailed explanation of the primary categories of data types in MySQL, along with examples.

## 1. Numeric Types

Numeric data types are used to store numbers.

### INTEGER (or INT)
- **Description**: A whole number without a fractional component.
- **Example**:
```sql
CREATE TABLE employees (
    employee_id INT,
    salary INT
);

### DECIMAL
- **Description**: A fixed-point precision number, suitable for storing monetary values.
- **Example**:
```sql
CREATE TABLE products (
    product_id INT,
    price DECIMAL(10,2)  -- Can store values up to 99999999.99
);

### FLOAT

- **Description**: A floating-point number with a decimal component.
- **Example**:
```sql
CREATE TABLE sensor_data (
    sensor_id INT,
    temperature FLOAT  -- Suitable for sensor readings that may have decimal values
);


## 2. String Types

String data types are used to store text and character data.


### CHAR

- **Description** : A fixed-length character string.
- **Example**:
```sql
CREATE TABLE customers (
    customer_id INT,
    customer_name CHAR(50)  -- Fixed length of 50 characters
);


### VARCHAR

- **Description**: A variable-length character string.
- **Example** :

```sql

CREATE TABLE orders (
    order_id INT,
    customer_name VARCHAR(50)  -- Variable length, up to 50 characters
);


### TEXT

- **Description**: A variable-length text string for large data.
- **Example** :
```sql

CREATE TABLE articles (
    article_id INT,
    article_content TEXT  -- Suitable for large text data like articles or blog posts
);


## 3. Date and Time Types:

Date and time data types are used to store date and time values.

### DATE :

- **Description**: Date value in 'YYYY-MM-DD' format.
- **Example**:

```sql
CREATE TABLE events (
    event_id INT,
    event_date DATE  -- Stores dates such as '2024-10-09'
);


### DATETIME :

- **Description**: Date and time value in 'YYYY-MM-DD HH:MM' format.
- **Example**:
```sql
CREATE TABLE log_entries (
    log_id INT,
    log_timestamp DATETIME  -- Stores both date and time
);


### TIMESTAMP

- __Description__: A timestamp representing the number of seconds since the epoch (1970-01-01 00:00:00).
- __Example__:
```sql
CREATE TABLE posts (
    post_id INT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP  -- Automatically sets the timestamp when a record is created
);


```sql
/*
1. Create a database  
2. Use the database  
3. Create a table  
4. Insert records into tables  
5. Operations and functions
*/

-- Create a database 
    
create database datascience12pm;

-- View the database
show databases;

-- We need to use the database
use datascience12pm;

-- Create a table
create table info(
  id int,
  full_name varchar(30),
  phone_number varchar(10)
);

-- View tables
show tables;

-- Insert the records
insert into info (id, full_name, phone_number)
values
  (1, "Deva", "9499334444"),
  (2, "Sakku", "7036989237");

-- View the records from table
select * from info;

### Example - 2
```sql
-- Create the database
CREATE DATABASE VamsiHospitals;

-- Show databases
SHOW DATABASES;

-- Use the created database
USE VamsiHospitals;

-- Create the HospitalRecords table
CREATE TABLE HospitalRecords (
    PatientID INT,
    PatientName VARCHAR(100),
    Age INT,
    Disease VARCHAR(100),
    AdmissionDate DATE
);

-- Insert records into the HospitalRecords table
INSERT INTO HospitalRecords (PatientID, PatientName, Age, Disease, AdmissionDate) VALUES 
(1, 'Gopal', 45, 'Pneumonia', '2024-10-10'),
(2, 'Nitish', 30, 'Diabetes', '2024-09-15'),
(3, 'Sakku', 25, 'Fracture', '2024-10-12'),
(4, 'Raju', 50, 'Hypertension', '2024-08-20'),
(5, 'Sandeep', 60, 'Heart Disease', '2024-07-01');

-- Select all records from the HospitalRecords table
SELECT * FROM HospitalRecords;


### SQL Commands for Database Operations

-- create a Database

-- Use the Database

-- Create a Table

-- Insert Values into Records 

```sql
-- 1. Create a Database

-- CREATE DATABASE database_name;

CREATE DATABASE deva;
SHOW DATABASES;

-- 2. Use the Database

--USE database_name

USE deva;

-- 3. Create a Table

-- CREATE table table_name (columns (datatypes));

CREATE TABLE info (
    id INT,
    full_name VARCHAR(20),
    salary INT
);

-- View the Table
SHOW TABLES;

-- 4. Insert Records

-- insert into table_name (columns) values (records);
INSERT INTO info (id, full_name, salary) 
VALUES 
(1, 'Vamsi', 10000),
(2, 'Gopal', 50000),
(3, 'Sakku', 100000);

-- View the Records
SELECT * FROM info;



-- Drop the Database

DROP DATABASE deva;

### Session - 2

``` sql
/*

SQL NULL Values
SQL Update Statements
SQL Delete Statements
SQL Alter Table

---- Add Column in Existing Column

---- Alter Table - Drop Column

DROP
DELETE
TRUNCATE
*/

--CREATE Databse

create database customer;
show databses;

-- use database

use customer;

-- create customer_info table

create table customer_info (
id integer auto_increment,
first_name varchar(50),
last_name varchar(50),
salary interger,
PRIMARY KEY(id)
);

show tables;     -- Primary keys in MySQL are used to uniquely identify each row in a table

-- insert records into tables

insert into customer_info(first_name, last_name, salary)
values
('Vamsi','Reddy','15000'),
('Siva','Sai','67000'),
('Krishna','Kumar','60000'),
('Vicky','Sai',NULL); -- It means that value is missing here.

show tables;

select * from customer_info;

-- Getting the all null values in table

select * from customer_info where salary in null;

-- SQL Update statement it used to Replace the null values.

update customer_info set salary = 70000 where id = 2;
select * from customer_info;

-- Delete Statement - Used to delete the data.

delete from customer_info where id=2;
select * from customer_info;

--Alter table main
-- Add the column using Alter to existing table

alter table customer_info add phonenumber interger;
select * from customer_info;

alter table custmoer_info add dob DATE;
select * from customer_info;

desc customer_info;

-- DROP the columb using alter

alter table customer_info drop column dob;
select * from customer_info;

```sql

create table employees (
employee_id int auto_increment PRIMARY KEY,
first_name varchar(50),
last_name varchar(50),
department varchar(50)
);

-- Inserting some records

insert into employee (first_name, last_name, department)
VALUES
('Rahul','Sharma','HR'),
('Priya','Iyer','Finance'),
('Amit','Patel','IT'),
('Sneha','Reddy','IT');

-- DELETE Example :- Deletes Specific rows or all rows but keep the based on id

Delete from employee where employee_id == 2;
select * from employee;

-- TRUNCATE Example :- Removes all rows,resets auto-increment, keeps the start records.

TRUNCATE table employee;
select * from employee;

-- DROP Example :- Deletes the table structure and all its data permanently.

DROP table employee;

## Difference between DELETE, TRUNCATE, and DROP in MySQL

### 1. DELETE
- **Removes specific rows** from a table.
- **Slower** because it processes each row and logs them.
- Can be **rolled back** if used in a transaction.
- Keeps the table structure (columns, indexes, etc.) intact.
  
  ```sql
  DELETE FROM table_name WHERE co
### 2. TRUNCATE
- **Removes all rows** from a table.
- **Faster** than DELETE because it doesn't log individual row deletions.
- Cannot be rolled back, it's **non-transactional**.
- Keeps the table structure (columns, indexes, etc.) intact but removes all data.

```sql
TRUNCATE TABLE table_name;
_name;
ndition;


### 3. DROP
- **Removes the entire table**, including both data and structure (columns, indexes, constraints).
- **Very fast** because it drops the whole table at once.
- Cannot be rolled back; the table is permanently gone.
- No data or structure is left after the operation.

```sql
DROP TABLE table_name;

### Session - 3

```sql

/*
What is SQL Constraints ?
- SQL constraints are used to specify rules for the data in a table.
- Constraints are used to limit the type of data that can go into a table.
- There are 6 types of Constraints
1. NOT NULL - A column cannot have a Null Value.
2. UNIQUE - All values in a column are different.
3. PRIMARY KEY - A Combination of a NOT NULL and UNIQUE.
4. FOREIGN KEY - Prevents action that would destroy links between tables.
5. DEFAULT - Sets a default value for a column if no value is specified.
6. CHECK - The values in a column satisfies a specific Condition.
*/

create database Person;
show databases;
use Person;

-- 1. Create NOT NULL Constraint.

create table Person_info(
id int not null,
First_name varchar(35) not null,
Last_name varchar(35) not null,
age int
);
select * from Person;

-- to view the schema
desc Person_info;  -- describe - To understand complete information about database or schema.

--records
insert into Person_info(id,First_name,Last_name,age)
values
(1,'sdf','H',null);

selct * from Person_info;

-- Use the modify method to do changes in table.
alter table Person_info modify age int not null;
(1,'sdf','H',34);


-- 2. UNIQUE Constraint

use Person;
create table student (
id int not null,
first_name varchar(34),
last_name varchar(23),
age int not null,
unique(id)
);

desc students;

-- insert the values
insert into student(id,first_name,last_name,age)
values
(1,'vamsi','reddy',24),
(2,'vamsi','reddy',24);

selct * from student;

-- make a unique for another column
alter table student add unique(first_name);

--adding two features as unique.
alter table student
add constraint uqi_student unique(age,first_name);
desc student;

-- can we drop the constraint
alter table student
drop index uqi_student;


-- 3.PRIMARY KEY Constraint

create tablr student2(
id int,
fname varchar(25) not null,
lname varchar(25),
age int,
primary key(id)
);

desc student2;

-- records

insert into student2 values(1,'h','d',34);
insert into student2 values(2,'sdf','dsdf',34);

select * from student2;

--drop the Primary Key

alter table student2 drop

### Operators

```sql


/*
Operators are symbols or keywords used to perform operations on values and expressions.
They are fundamental components in SQL Queries, allowing you to filter, compare, and manipulate data.
Here are some common types of operators in MySQL:
1. ARITHMETIC OPERATOR
2. COMPARISON OPERATOR
3. LOGICAL OPERATOR
4. CONCATENATION OPERATOR
5. LIKE OPERATOR
6. IN OPERATOR
7. IS NULL / IS NOT NULL
8. BETWEEN

 OPERATOR
*/

-- 1. ARI

THMETIC OPERATOR
DROP DATABASE IF EXISTS sum;
CREATE DATABASE sum;
USE sum;

CREATE TABLE numbers (
    a INT,
    b INT
);
INSERT INTO numbers(a, b) VALUES
(10, 5),
(15, 7),
(25, 7);

SELECT a, b,
    a + b AS sum_result,
    a - b AS difference_result,
    a * b AS product_result,
    a / b AS division_result,
    a 

% b AS modulus_result
FROM

 numbers;

-- 2. COMPARISON OPERATOR
DROP DATABASE IF EXISTS com;
CREATE DATABASE com;
USE com;

CREATE TABLE products (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(30),
    price INT,
    stock_quantity INT
);
INSERT INTO products VALUES
(1, 'Rice', 30, 40),
(2, 'Dal', 96, 47),
(3, 'Fruit', 70, 30);

-- Comparison Queries
SELECT * FROM products WHERE price = 30;
SELECT * FROM products WHERE product_id != 2;
SELECT * FROM products WHERE price < 90;
SELECT * FROM products WHERE stock_quantity > 40;
SELECT * FROM products WHERE price

 <= 60;
SELECT * FROM 

products WHERE stock_quantity >= 30;

-- 3. LOGICAL OPERATOR
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    salary INT,
    department VARCHAR(50)
);
INSERT INTO employees VALUES
(1, 'John', 'Doe', 5000, 'Sales'),
(2, 'Jane', 'Smith', 6000, 'Marketing'),
(3, 'Bob', 'Johnson', 5500, 'Sales'),
(4, 'Alice', 'Williams', 7000, 'Finance'),
(5, 'Charlie', 'Brown', 4800, 'Marketing');

-- Logical Queries
SELECT * FROM employees WHERE department = 'Sales' AND salary > 5000;
SELECT * FROM employees WHERE department = 'Sales'


 OR salary < 5000;
SELECT * 

FROM employees WHERE NOT department = 'Marketing';

-- 4. CONCATENATION OPERATOR
CREATE DATABASE IF NOT EXISTS example_database;
USE example_database;

CREATE TABLE IF NOT EXISTS employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    salary INT
);
INSERT INTO employees VALUES
(1, 'John', 'Dev', 5000),
(2, 'Sai', 'Smith', 6000),
(3, 'Bob', 'John


', 5500);

SELECT 

employee_id, CONCAT(first_name, ' ', last_name) AS full_name FROM employees;

-- 5. LIKE OPERATOR
CREATE TABLE IF NOT EXISTS ele_products (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(100)
);
INSERT INTO ele_products VALUES
(1, 'Mobile Phone'),
(2, 'Laptop'),
(3, 'Tablet'),
(4, 'Smartwatch'),
(5, 'Mobile TV'),
(6, 'TV Tablet');

-- LIKE Queries
SELECT * FROM ele_products WHERE product_name LIKE 'Mobile%';
SELECT * FROM ele_products WHERE product_name LIKE '%Phone%';
SELECT * FROM ele_products WHERE product_name LIKE '%Tablet';

-- 7. IS NULL / IS NOT NULL
CREATE TABLE IF NOT EXISTS students (
    student_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50)
);
INSERT INTO students VALUES
(1, 'Vamsi', 'Reddy'),
(2, 'Sai', NULL),
(3, 'Bob', 'John'),
(4, 'Kumar', NULL);

SELECT * FROM students WHERE last_name IS NULL;
SELECT * FROM students WHERE last_name IS NOT NULL;

-- 8. BETWEEN OPERATOR
CREATE TABLE IF NOT EXISTS products_val (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(100),
    price DECIMAL(10, 2)
);
INSERT INTO products_val VALUES
(1, 'Laptop', 800.00),
(2, 'Smartphone', 200.00),
(3, 'Tablet', 120.00),
(4, 'Headphones', 60.00),
(5, 'Camera', 150.00);

SELECT * FROM products_val WHERE price BETWEEN 50.00 AND 100.00;
SELECT * FROM products_val WHERE price NOT BETWEEN 50.00 AND 100.00;


### SQL AGGREGATE FUNCTIONS

```sql

/*
SQL aggregate functions are used to perform calculations on sets of values.
These functions operate on multiple rows of data and return a single value,
summarizing the information in the dataset.
*/

DROP DATABASE IF EXISTS functions;
CREATE DATABASE functions;
USE functions;

CREATE TABLE sales (
    sale_id INT PRIMARY KEY,
    product_name VARCHAR(50),
    sale_date DATE,
    quantity INT,
    unit_price DECIMAL(10, 2)
);

-- Inserting records
INSERT INTO sales VALUES
(1, 'Laptop', '2022-01-15', 3, 1200.00),
(2, 'Smartphone', '2022-01-20', 5, 599.99),
(3, 'Tablet', '2022-02-05', 2, 299.99),
(4, 'Headphones', '2022-02-15', 10, 499.99),
(5, 'Laptop', '2022-03-10', 2, 1300.00);

-- COUNT(): Counts the number of rows in a result set.
SELECT COUNT(*) AS total_sales FROM sales;

-- SUM(): Calculates the sum of a numeric column.
SELECT SUM(quantity) AS total_quantity, SUM(unit_price * quantity) AS total_revenue FROM sales;

-- AVG(): Calculates the average value of a numeric column.
SELECT AVG(unit_price) AS average_price FROM sales;

-- MIN() AND MAX(): Retrieve the minimum and maximum values from a column.
SELECT MIN(unit_price) AS min_price, MAX(unit_price) AS max_price FROM sales;

-- GROUP BY: Groups the result set by one or more columns.
SELECT product_name, SUM(quantity) AS total_quantity FROM sales GROUP BY product_name;

-- HAVING: Filters the results of a GROUP BY query based on a condition.
SELECT product_name, SUM(quantity) AS total_quantity FROM sales GROUP BY product_name HAVING total_quantity = 5;


## SQL String Functions

``` sql
/*
String Functions that allow you to manipulate and perform operations on character strings.
*/

-- Creating a table for employee data.

DROP TABLE IF EXISTS employees;
CREATE TABLE employees(
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    email VARCHAR(100),
    salary DECIMAL(10, 2)
);

-- Inserting sample data into the employees table.

INSERT INTO employees(employee_id, first_name, last_name, email, salary)
VALUES
    (1, 'John', 'Doe', '   john.doe@example.com ', 50000.00),
    (2, 'Jane', 'Smith', '   jane.smith@example.com ', 60000.00),
    (3, 'Bob', 'Johnson', ' bob.johnson@example.com  ', 55000.00);

SELECT * FROM employees;




-- FUNCTIONS



-- CONCAT(): Concatenates two or more strings.

SELECT first_name, last_name, CONCAT(first_name, ' ', last_name) AS full_name FROM employees;



-- LENGTH(): Returns the length of a string.

SELECT first_name, LENGTH(first_name) AS name_length FROM employees;



-- UPPER() and LOWER(): Converts a string to uppercase or lowercase.

SELECT first_name, UPPER(first_name) AS upper_name, LOWER(first_name) AS lower_name FROM employees;

SELECT last_name, UPPER(last_name) AS upper_name, LOWER(last_name) AS lower_name FROM employees;




-- LEFT() and RIGHT(): Retrieves a specified number of characters from the left or right of a string.

SELECT first_name, LEFT(first_name, 1) AS left_first_name, RIGHT(first_name, 3) AS right_first_name FROM employees;




-- TRIM(): Removes leading and trailing spaces from a string.

SELECT email, TRIM(email) AS trimmed_email FROM employees;

SELECT * FROM employees;




-- REPLACE(): Replaces occurrences of a specified substring with another substring.

SELECT email, REPLACE(email, '@example.com', '@company.com') AS new_email FROM employees;




-- SUBSTRING(): Extracts a substring from a string.

SELECT first_name, SUBSTRING(first_name, 1, 3) AS substring_name FROM employees;


### JOINS

```sql

/*
JOINS

Joins allow you to retrieve data from two or more related tables based on a common field.

Types of Joins

1. INNER JOIN: Returns records that have matching values in both tables.

2. LEFT JOIN (or LEFT OUTER JOIN): Returns all records from the left table, and the matched records from the right table.
                                    If there is no match, the result is NULL from the right side.

3. RIGHT JOIN (or RIGHT OUTER JOIN): Returns all records from the right table, and the matched records from the left table.
                                     If there is no match, the result is NULL from the left side.
									
4. FULL OUTER JOIN: Returns all records when there is a match in either left or right table.
                    MySQL does not support FULL OUTER JOIN directly, but it can be simulated.
*/

-- Create a new database
CREATE DATABASE jointopic;
SHOW DATABASES;
USE jointopic;

-- Create Employees table
CREATE TABLE Employees (
    EmployeeID INT,
    Name VARCHAR(50),
    DepartmentID INT,
    Salary DECIMAL(10,2)
);

-- Create Departments table
CREATE TABLE Departments(
    DepartmentID INT,
    DepartmentName VARCHAR(50)
);

-- Insert data into Employees table
INSERT INTO Employees (EmployeeID, Name, DepartmentID, Salary)
VALUES
    (1, 'Vamsi', 1, 55000),
    (2, 'Swayam', 2, 65000),
    (3, 'Divya', 3, 48000),
    (4, 'Shankar', NULL, 75000),
    (5, 'Ravi', 1, 52000);

-- Insert data into Departments table
INSERT INTO Departments (DepartmentID, DepartmentName)
VALUES
    (1, 'Sales'),


        (2, 'Marketing'),
    (3, 'HR'),
    (4, 'IT');

-- 1. INNER JOIN 

-- Retrieves employees who have a matching DepartmentID in both tables.
SELECT Employees.Name, Departments.DepartmentName
FROM Employees
INNER JOIN Depar


tments ON Employees.DepartmentID = Departments.DepartmentID;

-- 2. LEFT JOIN

-- Retrieves all employees and their departments (NULL for those without a department).
SELECT Employees.Name, Departments.DepartmentName
FROM Employees
LEFT JOI


N Departments ON Employees.DepartmentID = Departments.DepartmentID;

-- 3. RIGHT JOIN

-- Retrieves all departments and their employees (NULL for departments without employees).
SELECT Employees.Name, Departments.DepartmentName
FROM Employees
RI



GHT JOIN Departments ON Employees.DepartmentID = Departments.DepartmentID;

-- 4. FULL OUTER JOIN (Simulated)

-- (Simulated using UNION): Retrieves all employees and departments, showing NULLs where there is no match.
SELECT Employees.Name, Departments.DepartmentName
FROM Employees
LEFT JOIN Departments ON Employees.DepartmentID = Departments.DepartmentID

UNION

SELECT Employees.Name, Departments.DepartmentName
FROM Employees
RIGHT JOIN Departments ON Employees.DepartmentID = Departments.DepartmentID;


## QUERIES and SUB QUERIES

```sql

/*
QUERIES : Main statements used to retrieve data from one or more tables.

SUB QUERIES : Nested queries within a main query, used for more complex data retrieval tasks.
              SUB QUERIES can appear in various places like SELECT, WHERE, and FROM clauses to help refine the main query’s result.
*/

CREATE TABLE Customers (
    CustomerID INT PRIMARY KEY,
    CustomerName VARCHAR(50),
    Contact VARCHAR(15)
);

CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    OrderDate DATE,
    Amount DECIMAL(10, 2),
    CustomerID INT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);

INSERT INTO Customers (CustomerID, CustomerName, Contact)
VALUES 
    (1, 'Alice', '555-1234'),
    (2, 'Bob', '555-5678'),
    (3, 'Charlie', '555-8765'),
    (4, 'David', '555-4321');

INSERT INTO Orders (OrderID, OrderDate, Amount, CustomerID)
VALUES 
    (101, '2024-10-01', 150.00, 1),
    (102, '2024-10-05', 250.00, 2),
    (103, '2024-10-07', 300.00, 1),
    (104, '2024-10-10', 500.00, 3),

    
    (105, '2024-10-15', 450.00, 2);

-- QUERIES

-- Basic Query

-- 1. Retrieve All Orders:

SELECT * FROM Orders;

-- 2. Retrieve Customer Names with Orders Above 200:

SELECT CustomerName 
FROM Customers 
JOIN Orders ON Customers.CustomerID = Orders.CustomerID 
WHERE Amount > 200;

-- 3. Total Order Amount per Customer:

SELECT CustomerID, SUM(A


mount) AS TotalSpent 
FROM Orders 
GROUP BY CustomerID;

-- SUB QUERIES

/*
1. Subquery in the SELECT Clause

A subquery in the SELECT clause is used to calculate a value that will appear as a column in the final result. 
It allows for dynamic calculations for each row in the main query.
*/

SELECT CustomerName, 
       (SELECT SUM(Amount) 
        FROM Orders 
        WHERE Orders.CustomerID = Customers.CustomerID) AS TotalSpent
FROM Customers;

/*
2. Subquery in the WHERE Clause

Subqueries in the WHERE clause are often used to filter records in the main query based on values from another query.
These are the most common types of subqueries and are often used for comparisons.
*/

SELECT CustomerName
FROM Customers 
WHERE CustomerID IN (
    SELECT CustomerID 
    FROM Orders 
    WHERE Amount > (SELECT AVG(Amount) FROM Orders)
);

/*
3. Subquery in the FROM Clause

A subquery in the FROM clause is often called a derived table or inline view. 
It allows the results of the subquery to be treated as a table that the main query can work with.
*/

SELECT Customers.CustomerName, MaxOrders.MaxAmount 
FROM Customers 
JOIN (SELECT CustomerID, MAX(Amount) AS MaxAmount 
      FROM Orders 
      GROUP BY CustomerID) AS MaxOrders 
ON Customers.CustomerID = MaxOrders.CustomerID;


## CLAUSES

```sql

/*
In MySQL, CLAUSES are keywords used to specify conditions or filters on queries, helping to retrieve, update, or delete data more effectively.
Here are some commonly used clauses in MySQL along with examples:

- WHERE Clause
- ORDER BY Clause
- GROUP BY Clause
- HAVING Clause
- LIMIT Clause
*/



/*
1. SELECT Clause:
The SELECT clause is used to specify the columns that you want to retrieve from a table.
*/

SELECT CustomerName, OrderDate FROM Orders;





/*
2. WHERE Clause:
The WHERE clause filters records based on a specified condition.
It's commonly used with SELECT, UPDATE, and DELETE statements.
*/

SELECT * FROM Orders WHERE Amount > 200;






/*
3. ORDER BY Clause:
The ORDER BY clause sorts the result set by one or more columns, either in ascending (ASC) or descending (DESC) order.
*/

SELECT * FROM Orders ORDER BY Amount DESC;




/*
4. GROUP BY Clause:
The GROUP BY clause groups rows that have the same values in specified columns into summarizations of data.
It's often used with aggregate functions (e.g., COUNT, SUM, AVG).
*/

SELECT CustomerID, COUNT(OrderID) AS TotalOrders 
FROM Orders 
GROUP BY CustomerID;




/*
5. HAVING Clause:
The HAVING clause is used to filter groups based on aggregate values.
It is similar to the WHERE clause but is applied to the results of GROUP BY.
*/

SELECT CustomerID, SUM(Amount) AS TotalSpent 
FROM Orders 
GROUP BY CustomerID 
HAVING TotalSpent > 500;




/*
6. LIMIT Clause:
The LIMIT clause limits the number of rows returned by a query.
*/

SELECT * FROM Customers LIMIT 10;
